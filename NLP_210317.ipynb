```python
import re
# separate the keywords into each column
df['genre'] = df['tag'].str.findall('genre:([a-zA-Z\/a-zA-Z\-a-zA-Z]+)', flags=re.IGNORECASE).apply(lambda x: ', '.join(map(str, x)))
df['mood'] = df['tag'].str.findall('mood:([a-zA-Z\/a-zA-Z\-a-zA-Z]+)', flags=re.IGNORECASE).apply(lambda x: ', '.join(map(str, x)))
df['channel'] = df['tag'].str.findall('channel:([a-zA-Z\/a-zA-Z\-a-zA-Z]+)', flags=re.IGNORECASE).apply(lambda x: ', '.join(map(str, x)))
df['distributor'] = df['tag'].str.findall('distributor:([a-zA-Z\/a-zA-Z\-a-zA-Z]+)', flags=re.IGNORECASE).apply(lambda x: ', '.join(map(str, x)))
df['branded-vod'] = df['tag'].str.findall('branded-vod:([a-zA-Z\/a-zA-Z\-a-zA-Z]+)', flags=re.IGNORECASE).apply(lambda x: ', '.join(map(str, x)))

df
```

```python
import pandas as pd
import string
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.corpus import stopwords
lemmatizer = nltk.stem.WordNetLemmatizer()
```

```python
df['texts'] = df['description'] + ', ' + df['genre'] + ', ' + df['mood']
df
```

## plot clustering

```python
# code from seminar
def clean_text(x, stop_words):
  if pd.isna(x) == False:
    x = x.lower()
    x = x.translate(str.maketrans(' ', ' ', string.punctuation))
    x = x.strip()

    x = nltk.word_tokenize(x)
    x = [token for token in x if not token in stop_words]
    x = ' '.join([lemmatizer.lemmatize(w) for w in x])
  else:
    x = ''

  return x
```

```python
stop_words = stopwords.words('english')
df['tokens'] = df['texts'].apply(lambda x: clean_text(x, stop_words))
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn import cluster

vectorizer = TfidfVectorizer(min_df=2, max_df=0.8, norm='l2')
X = vectorizer.fit_transform(df['tokens'])
tf_idf = pd.DataFrame(data = X.toarray(), columns=vectorizer.get_feature_names())
tf_idf
```

```python
clusters = 7
kmeanModel = KMeans(n_clusters=clusters, init='k-means++', max_iter=3000, random_state=0)
mod = kmeanModel.fit_transform(tf_idf)
df['k_means'] = kmeanModel.predict(tf_idf)
```

```python
order_centroids = kmeanModel.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
dict = []
for i in range(clusters):
  print('%d' % i, sep='', end=', '),
  for ind in order_centroids[i, :20]:
    print(terms[ind], sep='', end=', ')
  print('')
```

```python
df['genre'] = df['genre'].str.replace('drama/','')
df
```

```python
vectorizer = TfidfVectorizer(min_df=2, max_df=0.8, norm='l2')
X = vectorizer.fit_transform(df['genre'] + ', ' + df['mood'])
tf_idf = pd.DataFrame(data = X.toarray(), columns=vectorizer.get_feature_names())
tf_idf
```

```python
clusters = 7
kmeanModel = KMeans(n_clusters=clusters, init='k-means++', max_iter=3000, random_state=0)
mod = kmeanModel.fit_transform(tf_idf)
df['k_genre'] = kmeanModel.predict(tf_idf)
```

```python
order_centroids = kmeanModel.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
dict = []
for i in range(clusters):
  print('%d' % i, sep='', end=', '),
  for ind in order_centroids[i, :20]:
    print(terms[ind], sep='', end=', ')
  print('')
```

```python
genre = []
for row in df['k_genre']:
    if row == 0 :    genre.append('Thriller')
    elif row == 1:   genre.append('Drama')
    elif row == 2:   genre.append('Real Life Story')
    elif row == 3:   genre.append('Romance')
    elif row == 4:   genre.append('Mystery')
    elif row == 5:   genre.append('Action')
    elif row == 6:   genre.append('Comedy')
    else:           genre.append('na')

df['GenreType'] = genre
df
```

```python
## Language
```

```python
lang = []
for row in df['description']:
    if row.find('German') >0 :    lang.append('German')
    elif row.find('Spanish') >0:   lang.append('Spanish')
    elif row.find('French') >0:   lang.append('French')
    elif row.find('Greek') >0:   lang.append('Greek')
    elif row.find('Mandarin') >0:   lang.append('Mandarin')
    elif row.find('Japanese') >0:   lang.append('Japanese')
    elif row.find('Bulgarian') >0:   lang.append('Bulgarian')
    else:           lang.append('English')

lang
df['Language'] = lang
df
```
